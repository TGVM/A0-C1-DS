### Importação dos dados



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


Nesta primeira análise, você deve calcular o faturamento total de cada loja somando os valores da coluna Preço de cada loja para estimar o faturamento.

```python
somatorio = 0.0
for preco in loja.Preço:
  somatorio += preco
print("Faturamento da loja 1: "+str(somatorio))

somatorio2 = 0.0
for preco in loja2.Preço:
  somatorio2 += preco
print("Faturamento da loja 2: "+str(somatorio2))

somatorio3 = 0.0
for preco in loja3.Preço:
  somatorio3 += preco
print("Faturamento da loja 3: "+str(somatorio3))

somatorio4 = 0.0
for preco in loja4.Preço:
  somatorio4 += preco
print("Faturamento da loja 4: "+str(somatorio4))
```

Gráfico comparando as 4 lojas:

```python
listaLojas = ["Loja1", "Loja2", "Loja3", "Loja4"]
listaFaturamento = [somatorio, somatorio2, somatorio3, somatorio4]
plt.figure(figsize=(10, 5))
plt.bar(listaLojas, listaFaturamento, color=['blue', 'green', 'red', 'purple'])
plt.title('Faturamento por Loja em milhões de R$')
plt.xlabel('Lojas')
plt.ylabel('Faturamento (R$)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```

# 2. Vendas por Categoria


Neste passo, deve calcular a quantidade de produtos vendidos por categoria em cada loja. A ideia é agrupar os dados por categoria e contar o número de vendas de cada tipo, mostrando as categoria mais populares de cada loja

```python
vendasCategoria1 = {}
vendasCategoria2 = {}
vendasCategoria3 = {}
vendasCategoria4 = {}
for venda in loja.iloc[:, 1]:
  if venda in vendasCategoria1:
    vendasCategoria1[venda] += 1
  else:
    vendasCategoria1[venda] = 1

for venda in loja2.iloc[:, 1]:
  if venda in vendasCategoria2:
    vendasCategoria2[venda] += 1
  else:
    vendasCategoria2[venda] = 1

for venda in loja3.iloc[:, 1]:
  if venda in vendasCategoria3:
    vendasCategoria3[venda] += 1
  else:
    vendasCategoria3[venda] = 1

for venda in loja4.iloc[:, 1]:
  if venda in vendasCategoria4:
    vendasCategoria4[venda] += 1
  else:
    vendasCategoria4[venda] = 1
```

```python
print("Loja 1:")
soma1 = 0
soma2 = 0
soma3 = 0
soma4 = 0
for item in vendasCategoria1:
  print(item+": "+str(vendasCategoria1[item]))
  soma1 += vendasCategoria1[item]
print("Total: "+str(soma1))

print("\nLoja 2:")
for item in vendasCategoria2:
  print(item+": "+str(vendasCategoria2[item]))
  soma2 += vendasCategoria2[item]
print("Total: "+str(soma2))

print("\nLoja 3:")
for item in vendasCategoria3:
  print(item+": "+str(vendasCategoria3[item]))
  soma3 += vendasCategoria3[item]
print("Total: "+str(soma3))

print("\nLoja 4:")
for item in vendasCategoria4:
  print(item+": "+str(vendasCategoria4[item]))
  soma4 += vendasCategoria4[item]
print("Total: "+str(soma4))

```

Gráfico comparando as 4 lojas:

```python
listaLojas = [vendasCategoria1, vendasCategoria2, vendasCategoria3, vendasCategoria4]
categorias = list(vendasCategoria1.keys())  # Pega os nomes das categorias
width = 0.2  # Largura das barras
x = range(len(categorias))

plt.figure(figsize=(12, 6))
for i, LojaId in enumerate(listaLojas):
    valores = list(listaLojas[i].values())
    plt.bar([p + width*i for p in x], valores, width, label=LojaId)

plt.title('Vendas por Categoria em Cada Loja')
plt.xlabel('Categorias')
plt.ylabel('Quantidade Vendida')
plt.xticks([p + width*1.5 for p in x], categorias)
#plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# 3. Média de Avaliação das Lojas

Neste passo, vamos calcular a média das avaliações dos clientes para cada loja. O objetivo é entender a satisfação dos clientes com os produtos vendidos.

```python
media1=0
media2=0
media3=0
media4=0
for avaliacao in loja.iloc[:, 7]:
  media1 += avaliacao
media1 = media1/len(loja.iloc[:, 7])
print("Média da loja 1: " + str(media1))

for avaliacao in loja2.iloc[:, 7]:
  media2 += avaliacao
media2 = media2/len(loja2.iloc[:, 7])
print("Média da loja 2: " + str(media2))

for avaliacao in loja3.iloc[:, 7]:
  media3 += avaliacao
media3 = media3/len(loja3.iloc[:, 7])
print("Média da loja 3: " + str(media3))

for avaliacao in loja4.iloc[:, 7]:
  media4 += avaliacao
media4 = media4/len(loja4.iloc[:, 7])
print("Média da loja 4: " + str(media4))
```

Gráfico comparando as 4 lojas:

```python
mediaAvaliacao = [media1, media2, media3, media4]
listaLojas = ["Loja1", "Loja2", "Loja3", "Loja4"]

plt.figure(figsize=(10, 5))
plt.plot(listaLojas, mediaAvaliacao, marker='o', linestyle='-', color='orange', linewidth=2, markersize=8)
plt.title('Média de Avaliação por Loja')
plt.xlabel('Lojas')
plt.ylabel('Média de Avaliação (0-5)')
plt.ylim(0, 5)
plt.grid(linestyle='--', alpha=0.7)
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

Neste passo, deve identificar quais produtos foram os mais vendidos e os menos vendidos em cada loja.  Ao final, visualize os resultados de forma que fique claro quais produtos se destacaram nas vendas de cada loja.

```python
produtos1 = {}
produtos2 = {}
produtos3 = {}
produtos4 = {}
for produto in loja.iloc[:, 0]:
  if produto in produtos1:
    produtos1[produto] += 1
  else:
    produtos1[produto] = 1

for produto in loja2.iloc[:, 0]:
  if produto in produtos2:
    produtos2[produto] += 1
  else:
    produtos2[produto] = 1

for produto in loja3.iloc[:, 0]:
  if produto in produtos3:
    produtos3[produto] += 1
  else:
    produtos3[produto] = 1

for produto in loja4.iloc[:, 0]:
  if produto in produtos4:
    produtos4[produto] += 1
  else:
    produtos4[produto] = 1
```

```python
maisVendas1 = ["item", 0]
menosVendas1 = ["item", 1000]
maisVendas2 = ["item", 0]
menosVendas2 = ["item", 1000]
maisVendas3 = ["item", 0]
menosVendas3 = ["item", 1000]
maisVendas4 = ["item", 0]
menosVendas4 = ["item", 1000]
for item in produtos1:
  if produtos1[item] > maisVendas1[1]:
    maisVendas1[0] = item
    maisVendas1[1] = produtos1[item]
  if produtos1[item] < menosVendas1[1]:
    menosVendas1[0] = item
    menosVendas1[1] = produtos1[item]

for item in produtos2:
  if produtos2[item] > maisVendas2[1]:
    maisVendas2[0] = item
    maisVendas2[1] = produtos2[item]
  if produtos2[item] < menosVendas2[1]:
    menosVendas2[0] = item
    menosVendas2[1] = produtos2[item]

for item in produtos3:
  if produtos3[item] > maisVendas3[1]:
    maisVendas3[0] = item
    maisVendas3[1] = produtos3[item]
  if produtos3[item] < menosVendas3[1]:
    menosVendas3[0] = item
    menosVendas3[1] = produtos3[item]

for item in produtos4:
  if produtos4[item] > maisVendas4[1]:
    maisVendas4[0] = item
    maisVendas4[1] = produtos4[item]
  if produtos4[item] < menosVendas4[1]:
    menosVendas4[0] = item
    menosVendas4[1] = produtos4[item]
```

```python
print("Item mais vendido da loja 1: "+maisVendas1[0]+", com "+str(maisVendas1[1])+" vendas")
print("Item menos vendido da loja 1: "+menosVendas1[0]+", com "+str(menosVendas1[1])+" vendas\n")

print("Item mais vendido da loja 2: "+maisVendas2[0]+", com "+str(maisVendas2[1])+" vendas")
print("Item menos vendido da loja 2: "+menosVendas2[0]+", com "+str(menosVendas2[1])+" vendas\n")

print("Item mais vendido da loja 3: "+maisVendas3[0]+", com "+str(maisVendas3[1])+" vendas")
print("Item menos vendido da loja 3: "+menosVendas3[0]+", com "+str(menosVendas3[1])+" vendas\n")

print("Item mais vendido da loja 4: "+maisVendas4[0]+", com "+str(maisVendas4[1])+" vendas")
print("Item menos vendido da loja 4: "+menosVendas4[0]+", com "+str(menosVendas4[1])+" vendas\n")
```

Gráfico comparando as 4 lojas:

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

mais_vendidos = [maisVendas1, maisVendas2, maisVendas3, maisVendas4]
menos_vendidos = [menosVendas1, menosVendas2, menosVendas3, menosVendas4]

# Produtos mais vendidos
ax1.barh([prod[0] for prod in mais_vendidos], [prod[1] for prod in mais_vendidos], color='green')
ax1.set_title('Produtos Mais Vendidos')
ax1.set_xlabel('Quantidade Vendida')
ax1.grid(axis='x', linestyle='--', alpha=0.7)

# Produtos menos vendidos
ax2.barh([prod[0] for prod in menos_vendidos], [prod[1] for prod in menos_vendidos], color='red')
ax2.set_title('Produtos Menos Vendidos')
ax2.set_xlabel('Quantidade Vendida')
ax2.grid(axis='x', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

# 5. Frete Médio por Loja


Neste passo, vamos calcular o custo médio de frete para cada loja. O objetivo é entender quanto, em média, está sendo gasto com frete para cada uma das lojas.

```python
freteMedio1 = 0
freteMedio2 = 0
freteMedio3 = 0
freteMedio4 = 0
for frete in loja.Frete:
  freteMedio1 += frete
freteMedio1 = freteMedio1/len(loja.Frete)
print("Frete médio da loja 1: " + str(freteMedio1))

for frete in loja2.Frete:
  freteMedio2 += frete
freteMedio2 = freteMedio2/len(loja2.Frete)
print("Frete médio da loja 2: " + str(freteMedio2))

for frete in loja3.Frete:
  freteMedio3 += frete
freteMedio3 = freteMedio3/len(loja3.Frete)
print("Frete médio da loja 3: " + str(freteMedio3))

for frete in loja4.Frete:
  freteMedio4 += frete
freteMedio4 = freteMedio4/len(loja4.Frete)
print("Frete médio da loja 4: " + str(freteMedio4))
```

Gráfico comparando as 4 lojas:

```python
freteMedio = [freteMedio1, freteMedio2, freteMedio3, freteMedio4]
listaLojas = ["Loja1", "Loja2", "Loja3", "Loja4"]

plt.figure(figsize=(8, 8))
plt.pie(freteMedio, labels=listaLojas, autopct='%1.1f%%', startangle=90, colors=['gold', 'lightcoral', 'lightskyblue', 'lightgreen'])
plt.title('Distribuição do Frete Médio por Loja')
plt.show()
```
